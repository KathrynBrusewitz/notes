### 1.4 Simple statistical test

Using a t-test, compare the median house value in locations that are
adjacent to the Charles River to locations that are not adjacent to
the Charles River. Do you reject the null hypothesis that prices are
equivalent?

```python
# enter your code here
isAdj = boston.query('CHAS > 0')[['MEDV']]
notAdj = boston.query('CHAS < 1')[['MEDV']]

sp.stats.ttest_ind(isAdj,notAdj)

# Two-sided T-Test
```

*Write your response here.*
The resulting p-value is 7.39062317e-05. Such a small p-value (typically â‰¤ 0.05) indicates strong evidence against the null hypothesis, so we reject the null hypothesis that prices are equivalent.

### 2.1: Histogram of housing prices

### 2.2: Scatter plot of housing prices and crime

Use matplotlib to create a scatter plot that shows the relationship
between the median value of the home (y-axis) and the per-capita crime

```python
# enter your code here
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.plot(boston[['CRIM']], boston[['MEDV']], 'o', alpha=0.5)
plt.ylabel('median housing price (in $1000\'s)')
plt.xlabel('per-capita crime rate')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

# correlation
pd.Series.corr(boston['CRIM'], boston['MEDV'])
```
