## 2.1 Create binary variables

```python
# 1. Create binary variables (dummies) for pclass: pc3, pc2, and pc1
def pclassBinary(row, expect):
    if(row['pclass'] == expect):
        return 1
    else:
        return 0

# Add binary variables for pclass
df['pc3'] = df.apply(lambda row: pclassBinary(row, 3), axis=1)
df['pc2'] = df.apply(lambda row: pclassBinary(row, 2), axis=1)
df['pc1'] = df.apply(lambda row: pclassBinary(row, 1), axis=1)

def ageBinary(row, minAge, maxAge):
    if(minAge <= row['age'] <= maxAge):
        return 1
    else:
        return 0

def oldAgeBinary(row, minAge):
    if(minAge <= row['age']):
        return 1
    else:
        return 0

# Add binary variables for age
df['age 0-10'] = df.apply(lambda row: ageBinary(row, 0, 10), axis=1)
df['age 11-17'] = df.apply(lambda row: ageBinary(row, 11, 17), axis=1)
df['age 18-45'] = df.apply(lambda row: ageBinary(row, 18, 45), axis=1)
df['age 46-59'] = df.apply(lambda row: ageBinary(row,46, 59), axis=1)
df['age 60-inf'] = df.apply(lambda row: oldAgeBinary(row, 60), axis=1)

# We already have binary variables for embarked C Q and S
# And we already have binary variables for female and male

print(df.columns)
```

```python
# Split data into training/testing groups (80/20% or so)
binaryVars = ['sex_male', 'embarked_S', 'embarked_C', 'embarked_Q', 'pc3', 'pc2',
       'pc1', 'age 0-10', 'age 11-17', 'age 18-45', 'age 46-59', 'age 60-inf']

x = df[binaryVars]
y = df['survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

## 2.2 Implement decision tree with binary variables

```python
# Functions

# x = proportion survived
# y = proportion drowned

def ent(x, y):
    return -1 * x * math.log2(x) - y * math.log2(y)

# e0 = parent entropy
# e1, e2 = children entropies
# t0 = parent total samples
# t1, t2 = children total samples

def IG(e0, e1, e2, t0, t1, t2):
    return e0 - (((t1/t0)*e1) + ((t2/t0)*e2))
```

### 2.2.1
Compute the entropy at the root, based on the total number of survivors and drowned. You will use this number to compute entropy gain below

```python
total = len(y_train)
pSurvived = (y_train == 1).sum()/total
pDrowned = (y_train == 0).sum()/total

rootEnt = ent(pSurvived, pDrowned)

print('Root Entropy: ', rootEnt) # 0.959472807639
```

Root Entropy:  0.959472807639

### 2.2.2
Repeat the following with each feature you consider:

1. Split the data according to 0/1 answer to this feature. 
2. Compute entropy on both subsets
3. Compute the weighted average entropy 
4. Compute the entropy gain.

```python
largestGain = 0
largestVar = ''

for var in binaryVars:
    
    # Split the data according to 0/1 answer to feature 'var'
    filter1 = (x_train[var] == 1)
    filter2 = (x_train[var] == 0)
    subset1 = y_train[filter1]
    subset2 = y_train[filter2]
    
    total = len(subset1)+len(subset2)
    
    # Compute entropy on both subsets
    subset1ProportionSurvived = (subset1 == 1).sum()/len(subset1)
    subset1ProportionDrowned = (subset1 == 0).sum()/len(subset1)
    subset2ProportionSurvived = (subset2 == 1).sum()/len(subset2)
    subset2ProportionDrowned = (subset2 == 0).sum()/len(subset2)
    
    subset1Entropy = ent(subset1ProportionSurvived, subset1ProportionDrowned)
    subset2Entropy = ent(subset2ProportionSurvived, subset2ProportionDrowned)
    
    print('subset 1 entropy: ', subset1Entropy)
    print('subset 2 entropy: ', subset2Entropy)
    
    # Compute entropy gain
    infoGain = IG(rootEnt, subset1Entropy, subset2Entropy, total, len(subset1), len(subset2))
    
    if (infoGain > largestGain):
        largestGain = infoGain
        largestVar = var
        
    print('info gain: ', infoGain)
    print('---')
    
print('--- Results ---')
print('Largest info gain: ', largestGain)
print('Variable: ', largestVar)
```

Variable:  sex_male

### 2.2.3
Now pick the feature that gives the largest entropy gain. This gives you the  first split (tree stump). Print the survived/non-survived percentage in both branches.

```python
filter1 = (x_train[largestVar] == 1)
filter2 = (x_train[largestVar] == 0)
subset1 = y_train[filter1]
subset2 = y_train[filter2]

subset1ProportionSurvived = (subset1 == 1).sum()/len(subset1)
subset1ProportionDrowned = (subset1 == 0).sum()/len(subset1)
subset2ProportionSurvived = (subset2 == 1).sum()/len(subset2)
subset2ProportionDrowned = (subset2 == 0).sum()/len(subset2)


print('subset 1 proportion survived: ', subset1ProportionSurvived)
print('subset 1 proportion drowned: ', subset1ProportionDrowned)
print('subset 2 proportion survived: ', subset2ProportionSurvived)
print('subset 2 proportion drowned: ', subset2ProportionDrowned)
```
